```python
import requests
from bs4 import BeautifulSoup as bs 
import pandas as pd
```

```python
url ='https://vivareal.com.br/venda/parana/curitiba/apartamento_residencial/?Pagina{}'
```

```python
i = 1
r = requests.get(url.format(i))
soup = bs(r.text)
```

```python
soup
```

```python
houses = soup.find_all('a', {'class':'property-card__content-link js-card-title'})
qtd_imoveis = float(soup.find('strong', {'class': 'results-summary__count'}).text.replace('.',''))
houses 
```

```python
qtd_imoveis
```

```python
house = houses[0]
```

```python


descricao= house.find('span', {'class': 'property-card__title'}).text.strip()
endereco = house.find('span', {'class': 'property-card__address'}).text.strip()
area = house.find('span', {'class': 'property-card__detail-area'}).text.strip()
quartos = house.find('li', {'class': 'property-card__detail-room'}).text.strip().replace('   ', ' - ')
wc = house.find('li', {'class': 'property-card__detail-bathroom'}).text.strip().replace('   ', ' - ')
vaga = house.find('li', {'class': 'property-card__detail-garage'}).text.strip().replace('   ', ' - ')
valor = house.find('div', {'class': 'property-card__price'}).text.strip().replace('   ', ' - ')



print(descricao + ',' + endereco + ', ' + area + ', ' +quartos + ', ' + wc + ' , ' +  vaga + ', ' +valor)

```

```python
df = pd.DataFrame(
    columns= [
            'descricao',
            'endereco',
            'area',
            'quartos',
            'wc',
            'vaga',
            'valor'
    ]
)
```

```python
for house in houses:    
    try:
        descricao= house.find('span', {'class': 'property-card__title'}).text.strip()
    except:
        descricao = None 
    try:
        endereco = house.find('span', {'class': 'property-card__address'}).text.strip()
    except:
        endereco = None
    try:
        area = house.find('span', {'class': 'property-card__detail-area'}).text.strip()
    except:
        area = None
    try:
        quartos = house.find('li', {'class': 'property-card__detail-room'}).text.strip().replace('   ', ' - ')
    except:
        quartos = None
    try:
        wc = house.find('li', {'class': 'property-card__detail-bathroom'}).text.strip().replace('   ', ' - ')
    except:
        wc = None
    try:
        vaga = house.find('li', {'class': 'property-card__detail-garage'}).text.strip().replace('   ', ' - ')
    except:
        vaga = None
    try:
        valor = hovalor = house.find('div', {'class': 'property-card__price'}).text.strip().replace('   ', ' -')
    except:
        valor = None
    df.loc[df.shape[0]] = [
            descricao,
            endereco,
            area,
            quartos,
            wc,
            vaga,
            valor
]    
```
